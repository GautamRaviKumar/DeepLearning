#### <span style="color:#0b486b"> Some vocabulary </span>

* In NumPy, each dimension is called an **axis**.
* The number of axes is called the **rank**.
    * For example, the above 3x4 matrix is an array of rank 2 (it is 2-dimensional). The first axis has length 3, the second has length 4.
    * For example, the above 2x3x4 3D array has rank 3 with axis 0 =2, axis 1 = 3, and axis 2 = 4. You might imagine this a a 3D rectangle with depth=2, height= 3, and width= 4.
* An array's list of axis lengths is called the **shape** of the array.
    * For example, the above matrix's shape is `(3, 4)`.
    * The rank is equal to the shape's length.
* The **size** of an array is the total number of elements, which is the product of all axis lengths (eg. 3*4=12)

```python
a = np.zeros((3,4))
a
```

### <span style="color:#0b486b">I.2 Reshaping an array </span>
#### `In place`
Changing the shape of an `ndarray` is as simple as setting its `shape` attribute. However, the array's size must remain the same.

```python
g = np.arange(24)
print(g)
print("Rank:", g.ndim)
```

```python
g.shape = (6, 4)
print(g)
print("Rank:", g.ndim)
```

```python
g.shape = (2, 3, 4)
print(g)
print("Rank:", g.ndim)
```

#### `reshape`
The `reshape` function returns a new `ndarray` object pointing at the *same* data. This means that modifying one array will also modify the other.

```python
g = np.arange(24)
g2 = g.reshape(4,6)
print(g)
print(g2)
print("Rank:", g2.ndim)
```

#### `ravel`
Finally, the `ravel` function returns a new one-dimensional `ndarray` that also points to the same data:

```python
g2.ravel()
```

### <span style="color:#0b486b">I.5 Arithmetic operations </span>
All the usual arithmetic operators (`+`, `-`, `*`, `/`, `//`, `**`, etc.) can be used with `ndarray`s. They apply *elementwise*:

```python
a = np.array([10, 12, 16, 15])
b = np.array([5,  4,  3,  2])
print("a + b  =", a + b)
print("a - b  =", a - b)
print("a * b  =", a * b)
print("a / b  =", a / b)
print("a // b  =", a // b)
print("a % b  =", a % b)
print("a ** b =", a ** b)
```

### <span style="color:#0b486b">I.6 Conditional operators </span>
The conditional operators also apply elementwise:

```python
m = np.array([20, -5, 30, 40])
m < [15, 16, 35, 36]
```

```python
m < 25  # equivalent to m < [25, 25, 25, 25]
```

```python
m[m < 25]
```

### <span style="color:#0b486b">I.8 Stacking arrays </span>
It is often useful to stack together different arrays. NumPy offers several functions to do just that. Let's start by creating a few arrays.

```python
q1 = np.full((3,4), 1.0)
q1
```

```python
q2 = np.full((4,4), 2.0)
q2
```

```python
q3 = np.full((3,4), 3.0)
q3
```

```python
q4 = np.vstack((q1, q2, q3))
q4
q4.shape
```

```python
q5 = np.hstack((q1, q3))
q5
q5.shape
```

```python
q7 = np.concatenate((q1, q2, q3), axis=0)  # Equivalent to vstack
q7
q7.shape
```

```python
q8 = np.stack((q1, q3))
print(q8)
print(q8.shape)
```

```python
r = np.arange(24).reshape(6,4)
r
```

```python
r1, r2, r3 = np.vsplit(r, 3)
print(r1)
print(r2)
print(r3)
```

```python
r4, r5 = np.hsplit(r, 2)
print(r4)
print(r5)
```

Gray-scale images are 2D tensors, while color images are 3D tensors. This section offers you a brief introduction of how to manipulate images using `OpenCV`.

```python
import cv2
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

We read an image from a file using `cv2.imread`. Note that `color_img` is a 3D tensor with the shape `(1112, 900, 3)` consisting of three chanels of `(1122, 900)`. By default in OpenCV, the first is the blue channel, the second is the green channel, and the last is the red channel. Subsequently, we use `cv2.cvtColor` to rearrange in the standard order of the red, blue, and green channels.

```python
color_img = cv2.imread('images/colorApple.jpg')
color_img = cv2.cvtColor(color_img, cv2.COLOR_BGR2RGB)
print(color_img.shape)
plt.imshow(color_img)
plt.show()
```

We convert the color image `color_img` a gray-scale image with the shape (1122, 900) and visualize it.

```python
gray_img = cv2.cvtColor(color_img, cv2.COLOR_RGB2GRAY)
print(gray_img.shape)
plt.imshow(gray_img)
plt.show()
```

We split `color_img` into three channels and then visualize them. Please pay attention to the order of the three channels.

```python
color_img = cv2.imread('images/colorApple.jpg')
(B_2D, G_2D, R_2D) = cv2.split(color_img)
print(R_2D.shape, G_2D.shape, B_2D.shape)
```

```python
plt.imshow(R_2D)
plt.show()
```

```python
plt.imshow(G_2D)
plt.show()
```

```python
plt.imshow(B_2D)
plt.show()
```

```python
zeros = np.zeros((color_img.shape[0],color_img.shape[1], 1), dtype="uint8")
```

```python
R_3D = np.expand_dims(R_2D,axis=2)
G_3D = np.expand_dims(G_2D,axis=2)
B_3D = np.expand_dims(B_2D,axis=2)
print(R_3D.shape, G_3D.shape, B_3D.shape)
```

We insert two zero channels to the single red channel in order and then visualize.

```python
R_3D_zeros = np.concatenate((R_3D, zeros, zeros), axis= 2)
print(R_3D_zeros.shape)
plt.imshow(R_3D_zeros)
plt.show()
```

We insert two zero channels to the single green channel in order and then visualize.

```python
G_3D_zeros = np.concatenate((zeros, G_3D, zeros), axis= 2)
print(G_3D_zeros.shape)
plt.imshow(G_3D_zeros)
plt.show()
```

We insert two zero channels to the single blue channel in order and then visualize.

```python
B_3D_zeros = np.concatenate((zeros, zeros, B_3D), axis= 2)
print(B_3D_zeros.shape)
plt.imshow(B_3D_zeros)
plt.show()
```

**<span style="color:red">Exercise 5</span>**: load an image, then add Gaussian noises to each channel, and visualize the blurry image. You can use a variable $\epsilon$ to declare a noise level. 

Hint: you can use the function `np.random.normal` to generate a 2D random array.

```python
import cv2
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
eps = 100
color_img = cv2.imread('images/colorApple.jpg')
(B_2D, G_2D, R_2D) = cv2.split(color_img)
B_2D_n = B_2D + np.random.normal(0, eps, B_2D.shape)
B_2D_n = np.expand_dims(B_2D_n, axis =2)
R_2D_n = R_2D + np.random.normal(0, eps, R_2D.shape)
R_2D_n = np.expand_dims(R_2D_n, axis =2)
G_2D_n = G_2D + np.random.normal(0, eps, G_2D.shape)
G_2D_n = np.expand_dims(G_2D_n, axis =2)
color_img_n = np.concatenate((R_2D_n, B_2D_n, G_2D_n), axis = 2)
plt.imshow(color_img_n/255.0)
plt.show()
```
